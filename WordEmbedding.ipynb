```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]

# Vocabulary size
vocab_size = 10000
```

```python
one_hot_repr = [one_hot(words,vocab_size) for words in sent]
one_hot_repr

# output shows the indexes of the words present in the sentence out of the total vocabulary size
```

```python
# Word Embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
embeded_doc = pad_sequences(one_hot_repr,padding='pre',maxlen=10)
embeded_doc
```

```python
# Feature Representation : Training entire model's embeding layer (10 dimensions) from scratch. eg. word2vec(300 dimensions)
model =  Sequential()
model.add(Embedding(vocab_size,10,input_length=10))
# model.add(Embedding(input_dim=vocab_size, output_dim=10, input_length=10))
model.compile('adam','mse')
model.summary()

```

```python
model.predict(embeded_doc)
```
